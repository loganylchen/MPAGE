```r
modules <- readRDS("inst/extdata/rmp_modules.rds")
exp_data <- readRDS("data-raw/GSE26378_GPL570_gene_expression.rds")
pheno_data <- readRDS("data-raw/GSE26378_GPL570_phenotype.rds")

names(modules) <- paste0("M_", 1:length(modules))
```

```r
sample_condition <- pheno_data %>%
    dplyr::select(`disease state:ch1`) %>%
    dplyr::mutate(condition = ifelse(`disease state:ch1` == "septic shock patient", "case", "control"))
sample_condition_list <- sample_condition$condition
names(sample_condition_list) <- rownames(sample_condition)
sample_condition_list
```

```r
module_genes <- lapply(modules, function(module) {
  intersect(module$genes, rownames(exp_data))
})

# Remove empty modules and modules with too few genes
min_genes <- 3
module_genes <- module_genes[sapply(module_genes, length) >= min_genes]
```
